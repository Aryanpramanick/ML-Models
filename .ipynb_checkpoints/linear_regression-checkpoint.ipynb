# Multiple Linear Regression
## Prediction of dud 
### Dud-An indicator of censoring.These are categories for which full information is not available. 

#### We will use a dataset that stores the following columns.


Year - The year of the diagnosis.


Quarter - The quarter of the year in which diagnosis was made.

Delay - The time delay (in months) between diagnosis and reporting. 0 means that the case was reported within one month. Longer delays are grouped in 3 month intervals and the value of delay is the midpoint of the interval (therefore a value of 2 indicates that reporting was delayed for between 1 and 3 months).


Dud - An indicator of censoring. These are categories for which full information is not yet available and the number recorded is a lower bound only.

Time - The time interval of the diagnosis. That is the number of quarters from July 1983 until the end of the quarter in which these cases were diagnosed.


Y - The number of AIDS cases reported.

```python
import pandas as pd
```

```python
dataset = pd.read_csv("aids.csv" )
```

We will now divide the given dataset into training data and test data.

Division of dataset in 80% as train data and 20% test data after shuffling.

```python
train_data = dataset.sample(frac=0.8)
train_data = train_data.sort_values("index").reset_index(drop=True)
print(train_data)
```

```python
test_data = pd.concat([dataset, train_data]).drop_duplicates(keep=False).reset_index(drop=True)
print(test_data)
```

```python
X_train = train_data[['year', 'quarter' , 'delay' , 'time' , 'y']].as_matrix()
Y_train = train_data[train_data.columns[4]].as_matrix()
print(X_train, Y_train)
```

```python
train_data[['year', 'quarter' , 'delay' , 'time' , 'y']]
```

```python
X_test = test_data[['year', 'quarter' , 'delay' , 'time' , 'y']].as_matrix()
Y_test = test_data[test_data.columns[4]].as_matrix()
print(X_test, Y_test)
index_test = test_data[test_data.columns[0]].as_matrix()
```

Simple Linear Regression Model
Y =  𝛽0  +  𝛽1𝑥  + ϵ

To create our model, we must "learn" or estimate the values of regression coefficients  𝛽0  and  𝛽1 . And once we've estimated these coefficients, we can use the model to predict responses. First we are going to use the Least Squares technique.

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, Y_train)
```

```python
y_pred = reg.predict(X_test)
print('Coefficients:  ', reg.coef_)
print("Mean squared error:  %.2f"
      % mean_squared_error(Y_test, y_pred))
print('Variance score:  %.2f' % r2_score(Y_test, y_pred))
```

### Plotting actual dud value against the predicted dud value when the predicted value is not rounded off.

```python
plt.scatter(index_test, Y_test,  color='black')

plt.scatter(index_test, y_pred, color='blue')

plt.xticks(())
plt.yticks(())

plt.xlabel('index')
plt.ylabel('ACTUAL(black) vs PREDICTED(blue)')

plt.show()
```

```python
y_pred = reg.predict(X_test).round()
print('Coefficients:  ', reg.coef_)
print("Mean squared error:  %.2f"
      % mean_squared_error(Y_test, y_pred))
print('Variance score:  %.2f' % r2_score(Y_test, y_pred))
```

### Since Dud is a binary value hence rounding off the predicted dud to reduce error and plotting a graph.

```python
plt.scatter(index_test, Y_test,  color='black')

plt.scatter(index_test, y_pred, color='blue')

plt.xticks(())
plt.yticks(())

plt.xlabel('index')
plt.ylabel('ACTUAL(black) vs PREDICTED(blue)')

plt.show()
```
